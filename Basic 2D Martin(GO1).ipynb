## Quadruped Model

When dealing with complex objects, we can generalize the process of calculating the COM by following these steps:

Gather Data: Start by collecting the necessary information about the object:

Mass of the object (m).
Inertia components: Ixx, Iyy, and Izz.
Geometry information, such as dimensions or shape details.
Decompose into Components: Understand the object's geometry and break it down into simpler components. These components could be points, lines, or surfaces, depending on the object's shape.

Calculate Component COMs: Calculate the center of mass for each individual component. For simple shapes like spheres, cylinders, and boxes, the center of mass can be easily calculated based on the geometric properties of the shape.

Apply Parallel Axis Theorem (if needed): If the components' COMs are not already aligned with the desired reference frame, use the parallel axis theorem to adjust the moment of inertia for each component to the desired reference frame.

Combine Component COMs: Calculate the total mass and the weighted sum of component COMs to find the overall COM of the object. The weighted sum takes into account both the mass and the position of each component.

Finalize COM Calculation: The final COM is the point in space where the object's mass can be considered to be concentrated for the purpose of calculating rotational motion or balance.

Using the example of the trunk link:


Given:
- Mass: 5.204 kg
- Inertia:
  - Ixx = 0.0168352186 kg*m^2
  - Iyy = 0.0656071082 kg*m^2
  - Izz = 0.0742720659 kg*m^2

We can use the parallel axis theorem to calculate the center of mass (COM) along each axis using the inertia components. The parallel axis theorem states that the moment of inertia about an axis parallel to an axis through the center of mass is the sum of the moment of inertia about the center of mass and the product of the mass and the square of the distance between the two axes.

The formula for the COM along each axis is:
COM = √(I / m)

Where:
- I is the moment of inertia along the axis
- m is the mass

calculate the COM along each axis:

1. **COM along the x-axis (COM_x):**
   I_x = Ixx + m * y^2 + m * z^2
   COM_x = √(I_x / m)
   COM_x = √(0.0168352186 + 5.204 * (0.04675^2 + 0.057^2) / 5.204) ≈ 0.299 m

2. **COM along the y-axis (COM_y):**
   I_y = Iyy + m * x^2 + m * z^2
   COM_y = √(I_y / m)
   COM_y = √(0.0656071082 + 5.204 * (0.1881^2 + 0.057^2) / 5.204) ≈ 0.062 m

3. **COM along the z-axis (COM_z):**
   I_z = Izz + m * x^2 + m * y^2
   COM_z = √(I_z / m)
   COM_z = √(0.0742720659 + 5.204 * (0.1881^2 + 0.04675^2) / 5.204) ≈ 0.080 m

So, the center of mass of the object, calculated using the provided mass and inertia components, is approximately:
COM = (0.299 m, 0.062 m, 0.080 m).

As fellow students, we've used the inertia components and mass to calculate the approximate center of mass along each axis.
**'NB'**: this calculation assumes a rigid object with uniform density.

```python
# DERIVE THE EOMs SYMBOLICALLY --------------------------------------------------------------------------------------------

# import libraries
import sympy as sym
import numpy as np
import pickle as pickle

sym.init_printing()
from IPython.display import display #for pretty printing

# create symbolic variables

# system parameters
g = sym.symbols('g')
mb,mf1,mf2,mh1,mh2,mt = sym.symbols(['m_{b}','m_{f1}','m_{f2}','m_{h1}','m_{h2}','m_{t}']) # mass
lb,lf1,lf2,lh1,lh2,lt = sym.symbols(['l_{b}','l_{f1}','l_{f2}','l_{h1}','l_{h2}','l_{t}']) # length
Inb,Inf1,Inf2,Inh1,Inh2,Int = sym.symbols(['I_{b}','I_{f1}','I_{f2}','I_{h1}','I_{h2}','I_{t}']) # mass

# generalized coordinates
sides = ['R']
x,y,thb,tht = sym.symbols(['x','y','\\theta_{b}','\\theta_{t}']) 
dx,dy,dthb,dtht = sym.symbols(['\dot{x}','\dot{y}','\dot{\\theta}_{b}','\dot{\\theta}_{t}']) 
ddx,ddy,ddthb,ddtht = sym.symbols(['\ddot{x}','\ddot{y}','\ddot{\\theta}_{b}','\ddot{\\theta}_{t}']) 

thf1  = [sym.symbols('\\theta_{f1%s}'%sides[i]) for i in range(len(sides))]
thf2  = [sym.symbols('\\theta_{f2%s}'%sides[i]) for i in range(len(sides))]
thh1  = [sym.symbols('\\theta_{h1%s}'%sides[i]) for i in range(len(sides))]
thh2  = [sym.symbols('\\theta_{h2%s}'%sides[i]) for i in range(len(sides))]

dthf1  = [sym.symbols('\dot{\\theta}_{f1%s}'%sides[i]) for i in range(len(sides))]
dthf2  = [sym.symbols('\dot{\\theta}_{f2%s}'%sides[i]) for i in range(len(sides))]
dthh1  = [sym.symbols('\dot{\\theta}_{h1%s}'%sides[i]) for i in range(len(sides))]
dthh2  = [sym.symbols('\dot{\\theta}_{h2%s}'%sides[i]) for i in range(len(sides))]

ddthf1  = [sym.symbols('\ddot{\\theta}_{f1%s}'%sides[i]) for i in range(len(sides))]
ddthf2  = [sym.symbols('\ddot{\\theta}_{f2%s}'%sides[i]) for i in range(len(sides))]
ddthh1  = [sym.symbols('\ddot{\\theta}_{h1%s}'%sides[i]) for i in range(len(sides))]
ddthh2  = [sym.symbols('\ddot{\\theta}_{h2%s}'%sides[i]) for i in range(len(sides))]

q = sym.Matrix([x]+[y]+[thb]+thf1+thf2+thh1+thh2)
dq = sym.Matrix([dx]+[dy]+[dthb]+dthf1+dthf2+dthh1+dthh2)
ddq = sym.Matrix([ddx]+[ddy]+[ddthb]+ddthf1+ddthf2+ddthh1+ddthh2)

# rotation matrix
def get_R(th_in):    
    R = sym.Matrix([[sym.cos(th_in),-sym.sin(th_in)],
                    [sym.sin(th_in),sym.cos(th_in)]])
        
    return R

# positions

# body
p_b = sym.Matrix([[x],[y],[thb]])
J_b = p_b.jacobian(q)
v_b = J_b*dq

R_b = get_R(thb)

p_shoulder = sym.Matrix([[x],[y]]) + R_b*sym.Matrix([[lb*dCOM['body']],[0]])
p_hip = sym.Matrix([[x],[y]]) + R_b*sym.Matrix([[-lb*(1-dCOM['body'])],[0]])

# legs
p_lf1 = []
p_lf2 = []
p_lh1 = []
p_lh2 = []

J_lf1 = []
J_lf2 = []
J_lh1 = []
J_lh2 = []

v_lf1 = []
v_lf2 = []
v_lh1 = []
v_lh2 = []

p_elbow = []
p_knee = []

p_footf = []
p_footh = []
v_footf = []
v_footh = []

for i in range(len(sides)):
    # foreleg
    R_lf1 = get_R(thf1[i])
    p_lf1.append(sym.Matrix([p_shoulder + R_lf1*sym.Matrix([[0],[-dCOM['front_thigh']*lf1]]),[thf1[i]]]))
    J_lf1.append(p_lf1[i].jacobian(q))
    v_lf1.append(J_lf1[i]*dq)
    
    p_elbow.append(p_shoulder + R_lf1*sym.Matrix([[0],[-lf1]])) 
    
    R_lf2 = get_R(thf2[i])
    p_lf2.append(sym.Matrix([p_elbow[i] + R_lf2*sym.Matrix([[0],[-dCOM['front_calf']*lf2]]),[thf2[i]]]))
    J_lf2.append(p_lf2[i].jacobian(q))
    v_lf2.append(J_lf2[i]*dq)
    
    p_footf.append(sym.Matrix([p_elbow[i] + R_lf2*sym.Matrix([[0],[-lf2]])]))
    v_footf.append(p_footf[i].jacobian(q)*dq)
    
    # hindleg
    R_lh1 = get_R(thh1[i])
    p_lh1.append(sym.Matrix([p_hip + R_lh1*sym.Matrix([[0],[-dCOM['rear_thigh']*lh1]]),[thh1[i]]]))
    J_lh1.append(p_lh1[i].jacobian(q))
    v_lh1.append(J_lh1[i]*dq)
    
    p_knee.append(p_hip + R_lh1*sym.Matrix([[0],[-lh1]]))
    
    R_lh2 = get_R(thh2[i])
    p_lh2.append(sym.Matrix([p_knee[i] + R_lh2*sym.Matrix([[0],[-dCOM['rear_calf']*lh2]]),[thh2[i]]]))
    J_lh2.append(p_lh2[i].jacobian(q))
    v_lh2.append(J_lh2[i]*dq)
    
    p_footh.append(sym.Matrix([p_knee[i] + R_lh2*sym.Matrix([[0],[-lh2]])]))
    v_footh.append(p_footh[i].jacobian(q)*dq)
    
# STEP 2: Energy --------------------------------------------------------------------------------------------------------
# mass matrices for each link
Mb = sym.Matrix([[mb,0,0],[0,mb,0],[0,0,Inb]])
Mf1 = sym.Matrix([[mf1,0,0],[0,mf1,0],[0,0,Inf1]])
Mf2 = sym.Matrix([[mf2,0,0],[0,mf2,0],[0,0,Inf2]])
Mh1 = sym.Matrix([[mh1,0,0],[0,mh1,0],[0,0,Inh1]])
Mh2 = sym.Matrix([[mh2,0,0],[0,mh2,0],[0,0,Inh2]])

T = 0.5*v_b.transpose()*Mb*v_b
V = mb*g*p_b[1] 

for i in range(len(sides)):
    T += 0.5*v_lf1[i].transpose()*Mf1*v_lf1[i] +\
    0.5*v_lf2[i].transpose()*Mf2*v_lf2[i] +\
    0.5*v_lh1[i].transpose()*Mh1*v_lh1[i] +\
    0.5*v_lh2[i].transpose()*Mh2*v_lh2[i]
    
    V += mf1*g*p_lf1[i][1] +\
    mf2*g*p_lf2[i][1] +\
    mh1*g*p_lh1[i][1] +\
    mh2*g*p_lh2[i][1]
    
# STEP 4: partials of T and V ----------------------------------------------------------------------------------------------
# term 1
Lg1 = sym.zeros(len(q),1)
for i in range(len(q)):
    dT_ddq = sym.Matrix([sym.diff(T,dq[i])]) # get partial of T in dq_i
    Lg1[i] = dT_ddq.jacobian(q)*dq + dT_ddq.jacobian(dq)*ddq #...then get time derivative of that partial

# term 3
Lg3 = sym.Matrix([T]).jacobian(q).transpose() # partial of T in q

# term 4
Lg4 = sym.Matrix([V]).jacobian(q).transpose() # partial of U in q

# STEP 5: generalized forces -----------------------------------------------------------------------------------------------

# forces
tauf1  = [sym.symbols('\\tau_{f1%s}'%sides[i]) for i in range(len(sides))]
tauf2  = [sym.symbols('\\tau_{f2%s}'%sides[i]) for i in range(len(sides))]
tauh1  = [sym.symbols('\\tau_{h1%s}'%sides[i]) for i in range(len(sides))]
tauh2  = [sym.symbols('\\tau_{h2%s}'%sides[i]) for i in range(len(sides))]

Gxf  = [sym.symbols('G_{xf%s}'%sides[i]) for i in range(len(sides))]
Gyf  = [sym.symbols('G_{yf%s}'%sides[i]) for i in range(len(sides))]
Gxh  = [sym.symbols('G_{xh%s}'%sides[i]) for i in range(len(sides))]
Gyh  = [sym.symbols('G_{yh%s}'%sides[i]) for i in range(len(sides))]

# joints -------------------------------------------------

# tail
tau_Q = sym.zeros(len(q),1)

for i in range(len(sides)):
    # shoulder
    tauf1_b = sym.Matrix([[0],[0],[-tauf1[i]]])
    tauf1_f1 = sym.Matrix([[0],[0],[tauf1[i]]])
    tau_Q += J_b.transpose()*tauf1_b + J_lf1[i].transpose()*tauf1_f1
    
    # elbow
    tauf2_f1 = sym.Matrix([[0],[0],[-tauf2[i]]])
    tauf2_f2 = sym.Matrix([[0],[0],[tauf2[i]]])
    tau_Q += J_lf1[i].transpose()*tauf2_f1 + J_lf2[i].transpose()*tauf2_f2
    
    # hip
    tauh1_b = sym.Matrix([[0],[0],[-tauh1[i]]])
    tauh1_h1 = sym.Matrix([[0],[0],[tauh1[i]]])
    tau_Q += J_b.transpose()*tauh1_b + J_lh1[i].transpose()*tauh1_h1
    
    # knee
    tauh2_h1 = sym.Matrix([[0],[0],[-tauh2[i]]])
    tauh2_h2 = sym.Matrix([[0],[0],[tauh2[i]]])
    tau_Q += J_lh1[i].transpose()*tauh2_h1 + J_lh2[i].transpose()*tauh2_h2

# GRF -------------------------------------------------------------
GRF_Q = sym.zeros(q.shape[0],q.shape[1])
for i in range(len(sides)):
    R_lf2 = get_R(thf2[i])
    GRF_r = R_lf2*sym.Matrix([[0],[-(1-dCOM['front_calf'])*lf2]])
    GRF_th = GRF_r[0]*Gyf[i] - GRF_r[1]*Gxf[i]
    GRF_Q += J_lf2[i].transpose()*sym.Matrix([[Gxf[i]],[Gyf[i]],[GRF_th]])
    
    R_lh2 = get_R(thh2[i])
    GRF_r = R_lh2*sym.Matrix([[0],[-(1-dCOM['rear_calf'])*lh2]])
    GRF_th = GRF_r[0]*Gyh[i] - GRF_r[1]*Gxh[i]
    GRF_Q += J_lh2[i].transpose()*sym.Matrix([[Gxh[i]],[Gyh[i]],[GRF_th]])
    
Q = tau_Q + GRF_Q

#AND combine!
EOM = Lg1 - Lg3 + Lg4 - Q 
```

```python
DOFs = [('x','b'),
        ('y','b'),
        ('thb','b'),
        ('thf1','R'),
        ('thf2','R'),
        ('thh1','R'),
        ('thh2','R')]
WDOFs = ['x','y'] 
sides = ['R']
legs = ['lf1','lf2','lh1','lh2']
links = ['b','lf1','lf2','lh1','lh2']
joints = ['shoulder','elbow','hip','knee']
contacts = [(c,s) for c in ['f','h'] for s in sides]
ground_constraints = ['contact','sliding_ps','sliding_ng']
signs = ['ps','ng']

def get_var_list(m,n):
    var_list = [m.g]+\
               [m.m[l] for l in links]+\
               [m.len[l] for l in links]+\
               [m.In[l] for l in links]+\
               [m.q[n,dof] for dof in DOFs]+\
               [m.dq[n,dof] for dof in DOFs]+\
               [m.ddq[n,dof] for dof in DOFs]+\
               [9.81*m.force_a[n,j,s] for j in joints for s in sides]+\
               [9.81*m.GRF[n,c,dof] for dof in WDOFs for c in contacts]
    return var_list

def create_model(n,hm,mu):
    m = ConcreteModel()

    # SETS-----------------------------------------------------------------------------------------------------------------------

    m.N = RangeSet(N) # finite elements
    m.L = Set(initialize = links) # link types

    m.J = Set(initialize = [(j,s) for j in joints for s in sides], dimen=2) # joints
    
    m.DOF = Set(initialize = DOFs, dimen = 2) # generalized coordinates

    m.WDOF = Set(initialize = WDOFs) # world frame coordinates
 
    m.C = Set(initialize = contacts, dimen=2) # ground contact points

    m.sgn = Set(initialize = signs) # sign set for positive and negative components
    
    m.ground_constraints = Set(initialize = ground_constraints) # set for indexing ground-contact-related penalties
    
    # PARAMETERS-----------------------------------------------------------------------------------------------------------------

    m.g = Param(initialize = 9.81)

    def get_m(n, l):
        if l == 'b':
            return mass['body']
        if l == 'lf1':
            return mass['front_thigh']
        if l == 'lf2':
            return mass['front_calf']
        if l == 'lh1':
            return mass['rear_thigh']
        if l == 'lh2':
            return mass['rear_calf']
        if l == 't':
            return mass['tail']
    m.m = Param(m.L, initialize = get_m) # mass of links

    def get_len(n, l):
        if l == 'b':
            return length['body']
        if l == 'lf1':
            return length['front_thigh']
        if l == 'lf2':
            return length['front_calf']
        if l == 'lh1':
            return length['rear_thigh']
        if l == 'lh2':
            return length['rear_calf']
        if l == 't':
            return length['tail']
    m.len = Param(m.L, initialize = get_len) # length of links

    def calculate_In(m, l): 
        if l == 'b':
            return inertia['body']
        if l == 'lf1':
            return inertia['front_thigh']
        if l == 'lf2':
            return inertia['front_calf']
        if l == 'lh1':
            return inertia['rear_thigh']
        if l == 'lh2':
            return inertia['rear_calf']
        if l == 't':
            return inertia['tail']
    m.In = Param(m.L, initialize = calculate_In) # moment of inertia

    # VARIABLES -----------------------------------------------------------------------------------------------------------------

    # system coordinates
    m.q = Var(m.N, m.DOF) # position
    m.dq = Var(m.N, m.DOF) # velocity
    m.ddq = Var(m.N, m.DOF) # acceleration
    
    m.force_a = Var(m.N, m.J) # actuator forces
    
    m.GRF = Var(m.N, m.C, m.WDOF) # ground reaction forces

    # bound variables
    for n in range(1,N+1):
        m.q[n,('y','b')].setlb(0)
        m.q[n,('thb','b')].setlb(-np.pi)
        m.q[n,('thb','b')].setub(np.pi)
        for c in contacts:
            m.GRF[n,c,'y'].setlb(0)
            
        
    # JOINT CONSTRAINTS ----------------------------------------------------------------------------------------------------------

    m.jointp = Var(m.N, m.J) # joint position, relative    
    m.jointy = Var(m.N, m.J, bounds = (0,None)) # joint height

    def def_jointp(m,n,j1,j2):
        j = (j1,j2)
        if j1 == 'tail':
            return m.jointp[n,j] == m.q[n,'tht',j2] - m.q[n,'thb','b']
        if j1 == 'shoulder':
            return m.jointp[n,j] == m.q[n,'thf1',j2] - m.q[n,'thb','b']
        if j1 == 'elbow':
            return m.jointp[n,j] == m.q[n,'thf2',j2] - m.q[n,'thf1',j2]
        if j1 == 'hip':
            return m.jointp[n,j] == m.q[n,'thh1',j2] - m.q[n,'thb','b']
        if j1 == 'knee':
            return m.jointp[n,j] == m.q[n,'thh2',j2] - m.q[n,'thh1',j2]
    m.def_jointp = Constraint(m.N, m.J, rule = def_jointp)

    def def_jointy(m,n,j1,j2):
        j = (j1,j2)
        var_list = get_var_list(m,n)
        return m.jointy[n,j] == lamb_jointy[j](*var_list)
    m.def_jointy = Constraint(m.N, m.J, rule = def_jointy)
        
# joint limits

    for n in range(1, N + 1):
        for s in sides:
            m.jointp[n, 'shoulder', s].setlb(-2.966)  # Closest approximation to -2.8
            m.jointp[n, 'shoulder', s].setub(0.6)  # Closest approximation to 0.6
            m.jointp[n, 'elbow', s].setlb(-0.2)  # Closest approximation to 0.8
            m.jointp[n, 'elbow', s].setub(2.7)  # Closest approximation to 2.7
            m.jointp[n, 'hip', s].setlb(-2.966)  # Closest approximation to -2.8
            m.jointp[n, 'hip', s].setub(0.6)  # Closest approximation to 0.6
            m.jointp[n, 'knee', s].setlb(-0.2 )  # Closest approximation to 0.8
            m.jointp[n, 'knee', s].setub(2.7)  # Closest approximation to 2.7
    
    # TIME AND INTEGRATION ---------------------------------------------------------------------------------------------------

    # variable timestep
    m.hm = Param(initialize = hm, mutable = True) # master timestep
    m.h = Var(m.N, bounds = (1,1.0000001))
    m.T = Param(initialize = N*hm*0.5)
    
    def time_sum(m):
        return sum([m.hm*m.h[n] for n in range(1,N+1)]) >= m.T
    m.time_sum = Constraint(rule = time_sum)

    # Numerical integration ---------------------------------------------------------------------------------------------------
    def integrate_p(m,n,dof1,dof2):
        if n == 1: return Constraint.Skip
        dof = (dof1,dof2)
        return m.q[n,dof] == m.q[n-1,dof] + m.hm*m.h[n]*m.dq[n,dof]
    m.integrate_p = Constraint(m.N, m.DOF, rule = integrate_p)

    def integrate_v(m,n,dof1,dof2):
        if n == 1: return Constraint.Skip
        dof = (dof1,dof2)
        return m.dq[n,dof] == m.dq[n-1,dof] + m.hm*m.h[n]*m.ddq[n,dof]
    m.integrate_v = Constraint(m.N, m.DOF, rule = integrate_v)
    
    # EOM --------------------------------------------------------------------------------------------------------------------
    
    def dynamics(m,n,dof1,dof2):
        dof = (dof1,dof2)
        if n == 1: return Constraint.Skip
        var_list = get_var_list(m,n)
        return lamb_EOM[dof](*var_list) == 0
    m.dynamics = Constraint(m.N, m.DOF, rule = dynamics)

    #$$ \tau = \tau_{ff} + k_p (p_{des} - p) + k_d (\omega_{des} - \omega) $$
    # motor model --------------------------------------------------------------------------------------------------------------
    max_speed = 30  # Maximum speed in rad/s
    #max_tau = 23.7  # Maximum torque in Nm
    #reduction_ratio = 6.33  # Reduction ratio

    # Calculate the corresponding maximum rotational speed at the motor shaft
    #max_motor_speed = max_speed * reduction_ratio

    #for n in range(1, N + 1):
        
                # Constrain torque within the maximum torque limits
                #m.tau[n, s, leg].setlb(-max_tau)
                #m.tau[n, s, leg].setub(max_tau)
                #print(tauf1_f1)

                # Constrain joint angular velocity within the maximum speed limits
                #m.dq[n, 'th_l1%s' % s].setlb(-max_motor_speed)
                #m.dq[n, 'th_l1%s' % s].setub(max_motor_speed)

            
    # GROUND MODEL -----------------------------------------------------------------------------------------------------------

    # ground parameters
    m.mu_k = Param(initialize = mu) # coefficient of kinetic friction
    m.mu_s = Param(initialize = 1.5*mu) # coefficient of static friction
    m.e = Param(initialize = 0.0) # coefficient of restitution
    
    # foot position and velocity -----------------------------------
    m.foot_height = Var(m.N, m.C, bounds = (0,None)) # foot position
    m.footv = Var(m.N, m.C, m.WDOF) # foot velocity
    m.footv2 = Var(m.N, m.C, m.WDOF, m.sgn, bounds = (0,None)) # foot velocity +ve and -ve parts

    def def_foot_height(m,n,c1,c2):
        c = (c1,c2)
        var_list = get_var_list(m,n)
        return m.foot_height[n,c] == lamb_pfoot[('y',c1,c2)](*var_list)
    m.def_foot_height = Constraint(m.N, m.C, rule = def_foot_height)

    def def_footv(m,n,c1,c2,dof):
        c = (c1,c2)
        if n == 1: return Constraint.Skip
        var_list = get_var_list(m,n)
        return m.footv[n,c,dof] == lamb_vfoot[(dof,c1,c2)](*var_list)
    m.def_footv = Constraint(m.N, m.C, m.WDOF, rule = def_footv)
    
    def def_footv2(m,n,c1,c2,dof):
        c = (c1,c2)
        if n == 1: return Constraint.Skip
        return m.footv[n,c,dof] == m.footv2[n,c,dof,'ps'] - m.footv2[n,c,dof,'ng']
    m.def_footv2 = Constraint(m.N, m.C, m.WDOF, rule = def_footv2)
    
    # friction -------------------------------------------------------------
    m.alpha = Var(m.N, m.C, bounds = (0,1))
    
    def friction(m,n,c1,c2):
        c = (c1,c2)
        if n == 1: return Constraint.Skip
        F = m.mu_k
        return m.GRF[n,c,'x'] == (1 - 2*m.alpha[n,c])*m.GRF[n,c,'y']*F
    m.friction = Constraint(m.N, m.C, rule = friction)
       
    # ground complementarity ------------------------------------------------------------------------------------------------
    m.ground_penalty = Var(m.N, m.C, m.ground_constraints, bounds = (0.0,None))

    def ground_complementarity(m,n,c1,c2,gc):
        if n == 1: return Constraint.Skip
        c = (c1,c2)
        
        if gc == 'contact':
            A = m.GRF[n,c,'y']
            if n < N:
                B = m.foot_height[n+1,c]
            else:
                B = m.foot_height[n,c]
                
        if gc == 'sliding_ps':
            A = 1 - m.alpha[n,c]
            B = m.footv2[n,c,'x','ps']
            
        if gc == 'sliding_ng':
            A = m.alpha[n,c]
            B = m.footv2[n,c,'x','ng']
        
        return A*B <= m.ground_penalty[n,c,gc]
    m.ground_complementarity = Constraint(m.N, m.C, m.ground_constraints, rule = ground_complementarity)
                                
    # OBJECTIVES -----------------------------------------------------------------------------------------------------------
    def MinPenalty(m):
        return sum([m.ground_penalty[n,c,gc] for n in range(1,N+1) for c in contacts for gc in ground_constraints])
    m.MinPenalty = Objective(rule = MinPenalty)

    return m
```

```python
def stand(m):
    for dof in DOFs:
        m.dq[1,dof].fix(0)
        m.q[1,dof].fix(0)
    m.q[1,'y','b'].unfix()
    for n in range(1,N+1):
        for c in contacts:
            m.foot_height[n,c].fix(0)                
    return m
    
```
